# Multi-arm Bandits
Bandit problems are a special case of reinforcement learning where there is only one state but many possible actions.

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
#np.seterr(divide='ignore', invalid='ignore')
#mpl.rcParams['figure.figsize'] = (18,6)
```

Let's create some bandits at random, each with random mean and standard deviation between [0,1).

```python
class Bandit(object):
    def __init__(self):
        self.mean = np.random.normal(scale=1)
        self.stddev = 1
    def pull(self):
        return np.random.normal(loc=self.mean, scale=self.stddev)
    def expected(self):
        return self.mean
```

```python
class Environment(object):
    def __init__(self, n):
        self.n = n
        self.bandits = [Bandit() for i in range(n)]
        self.expected = np.array([b.expected() for b in self.bandits])
        self.optimal = np.argmax(self.expected)
        self.names = [chr(ord('a')+i) for i in range(n)]
    def action(self, a):
        return self.bandits[a].pull()
    def plot(self):
        data = [[self.action(a) for i in range(1000)] for a in range(self.n)]
        plt.violinplot(data, range(self.n), showextrema=False, showmeans=True)
        plt.xticks(range(self.n), self.names)
        plt.show()
    def plot_Q(self):
        plt.bar(range(self.n), self.expected)
        plt.show()
```

```python
class Agent(object):
    def __init__(self, n, epsilon):
        self.n = n
        self.Q = np.zeros(n)
        self.counts = np.zeros(n)
        self.epsilon = epsilon
    def action(self):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.n)
        return np.random.choice(np.flatnonzero(self.Q == np.max(self.Q)))
    def update(self, action, reward):
        self.counts[action] += 1
        self.Q[action] += (reward - self.Q[action]) / self.counts[action]
    def plot_Q(self):
        plt.bar(range(self.n), self.Q)
        plt.show()
    def plot_action(self):
        counts = np.zeros(self.n)
        for i in range(1000):
            counts[self.action()] += 1
        plt.bar(range(self.n), counts, tick_label=range(self.n))
        plt.show()
```

```python
class Run(object):
    def __init__(self, env, agents):
        self.env = env
        self.agents = agents
        self.n_agents = len(agents)
    def run(self, count):
        self.count = count
        self.hist_actions = np.zeros((self.n_agents, count))
        self.hist_rewards = np.zeros((self.n_agents, count))
        self.hist_average = np.zeros((self.n_agents, count))
        self.hist_correct = np.zeros((self.n_agents, count))
        self.hist_optimal = np.zeros((self.n_agents, count))
        for t in range(count):
            for i, agent in enumerate(self.agents):
                a = agent.action()
                r = self.env.action(a)
                agent.update(a, r)
                self.hist_actions[i][t] = a
                self.hist_rewards[i][t] = r
                self.hist_average[i][t] = self.hist_rewards[i].sum() / (t + 1)
                self.hist_correct[i][t] = (a == env.optimal)
                self.hist_optimal[i][t] = self.hist_correct[i].sum() / (t + 1)
    def plot_rewards(self):
        means = [b.mean for b in self.env.bandits]
        plt.plot(np.transpose(self.hist_average))
        plt.legend(range(self.n_agents), loc=6, bbox_to_anchor=(1, 0.5))
        plt.ylim((sum(means)/len(means),max(means)*1.1))
        plt.show()
    def plot_actions(self):
        plt.plot(np.transpose(self.hist_actions), ',')
        plt.legend(range(self.n_agents), loc=6, bbox_to_anchor=(1, 0.5))
        plt.ylim((-1, self.env.n + 1))
        plt.show()
    def plot_optimal(self):
        plt.plot(np.transpose(self.hist_optimal))
        plt.legend(range(self.n_agents), loc=6, bbox_to_anchor=(1, 0.5))
        plt.show()
```

```python
num_bandits = 10
env = Environment(num_bandits)
epsilons = [0.0, 0.1, 0.01, 0.001, 0.0001]
agents = [Agent(num_bandits, e) for e in epsilons]
run = Run(env, agents)
run.run(2000)
```

```python
env.plot()
```

```python
run.plot_rewards()
```

```python
run.plot_optimal()
```
